```python
from utils.utils import load_json
from matplotlib import pyplot as plt
import numpy as np
```

```python
test_data = iter(load_json('data/test.json'))
```

```python
plt.figure(figsize=(16,12))
for i in range(10):
    plt.subplot(3,4,i+1)
    label = None
    img = None
    while label != i:
        t = next(test_data)
        label = t[1]
        img = t[0]
    img = np.array(img).reshape(32,-1)
    plt.imshow(img, plt.cm.gray)
    plt.title(f"Class = {label}")
```
